### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
from sympy import *
import numpy as np
```

```python
# описывем функцию
x = Symbol('x')
f = x*5 + 4*sin(2*x) + cos(3*x+3)
print (f)
```

```python
# вычисляем значение первой и второй производной функции
f1 = lambdify(x, f)
print ('first derivative в точке 1: ', f1(1))
t = f.diff(x)
print ('сама функция ', t)
f2 = lambdify(x, t)
print ('second derivative в точке 1: ', f2(1))
t2 = t.diff(x)
print ('сама функция ', t2)
```

### Задание 2

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt
#находим функцию 1 производной
x = Symbol('x')
f = (sin(2*x+1))**5
f1 = f.diff(x)
f2 = f1.diff(x)
print ('функция 1 производной ', f1)
print ('функция 2 производной ', f2)
```

```python
def pr_func(function, a, b):
    interval = np.arange(a, b+1, 0.01)
    image = [function.subs(x, value) for value in interval]
    fig = plt.figure(figsize=(10, 10))
    plt.plot(interval, image)
    plt.show()
```

```python
# строим график для 1 производной
pr_func(f1, -5,5)
```

```python
# строим график для 2 производной
pr_func(f2, -5, 5)
```
